```python
import os
import cv2
import mediapipe as mp
import numpy as np
import csv

src_base = "../data/asl_dataset/asl_dataset/"
output_csv = "data/vectors/mediapipe_vectors.csv"

mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode=True)

no_hand_images = []

def process_folder(src_folder, folder_name, writer):
    for img_name in os.listdir(src_folder):
        src_img_path = os.path.join(src_folder, img_name)
        if not img_name.lower().endswith(('.jpg', '.jpeg', '.png')):
            continue
        img = cv2.imread(src_img_path)
        if img is None:
            print(f"Image non trouvée ou corrompue: {src_img_path}")
            continue
        img = cv2.resize(img, (400, 400))
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        results = hands.process(img_rgb)
        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                vecteurs = []
                for lm in hand_landmarks.landmark:
                    vecteurs.extend([lm.x, lm.y, lm.z])
                # Ajoute le label (nom du dossier) + nom de l'image + vecteur
                writer.writerow([folder_name, img_name] + vecteurs)
        else:
            print(f"Aucune main détectée dans: {src_img_path}")
            no_hand_images.append(src_img_path)

with open(output_csv, "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    # En-tête : label + nom + 63 colonnes
    header = ["label", "filename"] + [f"{coord}_{i}" for i in range(21) for coord in ["x", "y", "z"]]
    writer.writerow(header)
    for folder in os.listdir(src_base):
        src_folder = os.path.join(src_base, folder)
        if os.path.isdir(src_folder):
            process_folder(src_folder, folder, writer)
        else:
            # Si c'est un fichier image dans le dossier racine
            process_folder(src_base, "", writer)
            break

print(f"Extraction terminée. Vecteurs sauvegardés dans : {output_csv}")
```
